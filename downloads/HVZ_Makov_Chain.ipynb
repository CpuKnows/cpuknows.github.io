```python
import sklearn
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
markov_matrix = pd.DataFrame(data=[[.5, .15, .0], [.25, .85, .05], [.25, .0, .95]], columns=["H", "Z", "V"], index=["H", "Z", "V"])

population = []
time_steps = 100
for t in np.arange(1, time_steps + 1):
    population.append(np.dot(markov_matrix, np.array([t, t, t])))

population = pd.DataFrame(population)

print(population.irow(-1))

plt.plot(np.arange(1, time_steps + 1), population.ix[:, 0], label="Human")
plt.plot(np.arange(1, time_steps + 1), population.ix[:, 1], label="Zombie")
plt.plot(np.arange(1, time_steps + 1), population.ix[:, 2], label="Vampire")
plt.legend(loc=2)
```

```python
from numpy import random

h_pop, z_pop, v_pop = 300, 0, 0
h_history = [h_pop]
z_history = [z_pop]
v_history = [v_pop]

time_steps = 1000

for t in np.arange(time_steps):
    h_roll, z_roll, v_roll = random.random(3)
    
    #human state
    if h_roll < markov_matrix['H']['H']:
        #Stays human
        pass
    elif markov_matrix['H']['H'] <= h_roll < (markov_matrix['H']['H'] + markov_matrix['H']['Z']):
        if h_pop:
            #H -> Z
            h_pop -= 1
            z_pop += 1
    else:
        if h_pop:
            #H -> V
            h_pop -= 1
            v_pop += 1
        
    #zombie state
    if z_roll < markov_matrix['Z']['H']:
        if z_pop:
            #Z -> H
            z_pop -= 1
            h_pop += 1
    else:
        #Stays zombie
        pass
        
    #vampire state
    if h_roll < markov_matrix['V']['V']:
        #Stays vampire
        pass
    else:
        if v_pop:
            #V -> Z
            v_pop -= 1
            z_pop += 1
    
    h_history.append(h_pop)
    z_history.append(z_pop)
    v_history.append(v_pop)
    
h_history = pd.DataFrame(h_history)
z_history = pd.DataFrame(z_history)
v_history = pd.DataFrame(v_history)

plt.plot(np.arange(time_steps + 1), h_history, label="Human")
plt.plot(np.arange(time_steps + 1), z_history, label="Zombie")
plt.plot(np.arange(time_steps + 1), v_history, label="Vampire")
plt.legend(loc=3)
```

```python
from sklearn import hmm

startprob = np.array([1.0, .0, .0])
transmat = np.array([[.5, .25, .25],
                     [.15, .85, .0],
                     [.0, .05, .95]])

model = hmm.GaussianHMM(n_components=3, covariance_type='full', startprob=startprob, transmat=transmat)

# The means of each component
# x-axis = desire to eat humans
# y-axis = preferencce for night time
means = np.array([[-1.0,  3.0], # H
                  [9.0, 5.0], # Z
                  [8.0, 9.0], # V
                  ])
# The covariance of each component
covars = [.5 * np.identity(2),
          .1 * np.identity(2),
          .8 * np.identity(2)]

# Instead of fitting it from the data, we directly set the estimated
# parameters, the means and covariance of the components
model.means_ = means
model.covars_ = covars

# Generate samples
X, Z = model.sample(100)

# Plot the sampled data
plt.plot(X[:, 0], X[:, 1], "-o", label="observations", ms=6, mfc="orange", alpha=0.7)
plt.xlabel("Desire to eat humans")
plt.ylabel("Preference for night")
plt.show()
```

```python
n_components = 3

# make an HMM instance and execute fit
model = hmm.GaussianHMM(n_components, covariance_type="diag", n_iter=1000)
model.fit([X])

# predict the optimal sequence of internal hidden state
hidden_states = model.predict(X)

print("Transition matrix")
print(model.transmat_)
print()

print("means and vars of each hidden state")
for i in range(n_components):
    print("%dth hidden state" % i)
    print("mean = ", model.means_[i])
    print("var = ", np.diag(model.covars_[i]))
    print()
```
