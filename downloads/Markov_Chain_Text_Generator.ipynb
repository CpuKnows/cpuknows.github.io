```python
def read_in_corpus(path, corpus_storage=None, degree=2):
    file_in = open(path, 'r')
    
    if corpus_storage is None:
        corpus_storage = dict()
    
    sentence_start = []
    word_list = file_in.read().split()
    sentence_start.append((word_list[0], word_list[1]))
    
    for i in range(len(word_list) - degree):
        key = tuple([word_list[i + x] for x in range(degree)])
        if key in corpus_storage:
            corpus_storage[key].append(word_list[i + degree])
        else:
            corpus_storage[key] = [word_list[i + degree]]
            
        if '.' in word_list[i + degree] and i < len(word_list) - 2:
            sentence_start.append((word_list[i + degree + 1], word_list[i + degree + 2]))
                                  
    sentence_start = tuple(sentence_start)
    
    file_in.close()
    return (corpus_storage, sentence_start)
```

```python
def generate_text(corpus_storage, sentence_start, length):
    from random import randint
    
    sentence_count = 0
    open_char = False
    single_quote = False
    double_quote = False
    parenthesis = False
    bracket = False
    curly_bracket = False
    
    seed_index = randint(0, len(sentence_start) - 1)
    seed_text = sentence_start[seed_index]
    
    result = []
    result.append(seed_text[0])
    result.append(seed_text[1])
    key = seed_text
    
    while True:
        if key[0][0] is '\'':
            open_char = True
            single_quote = True
        elif key[0][0] is '\"':
            open_char = True
            double_quote = True
        elif key[0][0] is '(':
            open_char = True
            parenthesis = True
        elif key[0][0] is '[':
            open_char = True
            bracket = True
        elif key[0][0] is '{':
            open_char = True
            curly_brace = True
            
        text_choices = corpus_storage[key]
        
        if open_char == True:
            for choice in text_choices:
                if single_quote == True and '\'' in choice[-2:]:
                    single_quote = False
                    open_char = False
                    next_state = choice
                if double_quote == True and '\"' in choice[-2:]:
                    double_quote = False
                    open_char = False
                    next_state = choice
                if parenthesis == True and ')' in choice:
                    parenthesis = False
                    open_char = False
                    next_state = choice
                if bracket == True and ']' in choice:
                    bracket = False
                    open_char = False
                    next_state = choice
                if curly_brace == True and '}' in choice:
                    curly_brace = False
                    open_char = False
                    next_state = choice
        else:
            next_state = corpus_storage[key][randint(0, len(text_choices) - 1)]
            
        result.append(next_state)
        
        if ('.' in next_state or 
            '?' in next_state or
            '!' in next_state):
            sentence_count += 1
            
            if sentence_count >= length:
                break
            
            seed_index = randint(0, len(sentence_start) - 1)
            seed_text = sentence_start[seed_index]
            result.append(seed_text[0])
            result.append(seed_text[1])
            key = seed_text
        else:
            key = (key[1], next_state)
        
    return result
    
```

```python
def to_story(word_list):
    story_string = ""
    
    for word in word_list:
        story_string = story_string + " " + word
        
    return story_string
```

```python
(storage, sentence_start) = read_in_corpus('D:\\Users\\john\\My Documents\\Markov Chain Text Generator\\Alice in Wonderland - Lewis Carol.txt')
story = generate_text(storage, sentence_start, 3)
with open('D:\\Users\\john\\My Documents\\Markov Chain Text Generator\\out.txt', 'w') as file_out:
    file_out.write(to_story(story))
```
